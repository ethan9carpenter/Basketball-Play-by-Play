```python
import pandas as pd
import sqlite3 as sql
from collections import Counter, OrderedDict
```

```python
conn = sql.connect("nba_extended.db")
cur = conn.cursor()
df = pd.read_sql("""
                 SELECT
                     *
                 FROM
                     "find_names"
                     
                 """, conn)
pairsDF = Counter(list(zip(df['EVENTMSGACTIONTYPE'], df['EVENTMSGTYPE'])))
pairsDF = OrderedDict(dict(pairsDF))
pairs = list(pairsDF.keys())
action = [pair[0] for pair in pairs]
typ = [pair[1] for pair in pairs]
pairsDF = pd.DataFrame({'EVENTMSGACTIONTYPE': action, 'EVENTMSGTYPE': typ, 'Count': list(pairsDF.values())})
pairsDF.sort_values(['EVENTMSGACTIONTYPE', 'EVENTMSGTYPE'], inplace=True)
```

```python
pairsDF = pairsDF[['EVENTMSGACTIONTYPE', 'EVENTMSGTYPE']]
playsDF = []
df = df.sample(frac=1).reset_index(drop=True)
```

```python
for action, typ in zip(pairsDF['EVENTMSGACTIONTYPE'], pairsDF['EVENTMSGTYPE']):
    
    plays = df[(df['EVENTMSGACTIONTYPE'] == action) & (df['EVENTMSGTYPE'] == typ)][:1000]
    #playsDF.append(plays)
    plays.to_csv('play_numbers/{}-{}-({}).csv'.format(action, typ, len(plays)))
```

```python
playsDF = pd.concat(playsDF)
```
