```python
from tempAnalysisPrep import prep
from collections import Counter
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect("ncaa_pbp.db")
year = 2014
df = pd.read_sql("""
                    SELECT
                        EventPlayerID, EventTeamID, EventType, EventID
                    From
                        "poss_{}"
                    WHERE
                        EventType NOT LIKE "%8"
                     """.format(year), conn, index_col='EventID')
df = prep(df)
```

```python
data = df[df['Count'] >= 25]
data = data.sort_values('AveragePoints', ascending=False)
```
