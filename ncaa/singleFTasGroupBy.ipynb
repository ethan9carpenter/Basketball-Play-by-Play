```python
import sqlite3 as sql
import pandas as pd
```

```python
df = pd.read_sql("""Select * from 
                    "2017-2018" WHERE
                        EventType NOT LIKE "%8"
                        AND EventType IN 
                            (SELECT EventType FROM events WHERE isOffense = 1)
                            Limit 10000
                    """, sql.connect('ncaa_pbp.db'))
```

```python
df = df[df['EventType'] != 'assist']
df = df[df['EventType'].isin(['made1_free', 'miss1_free'])]
```

```python
group = df.groupby(['ElapsedSeconds', 'WTeamID', 'LTeamID', 'DayNum', 'WTeamID', 'EventPlayerID', 'EventType'])
```
