```python
import sqlite3 as sql
import pandas as pd
```

```python
conn = sql.connect('ncaa_pbp.db')
```

```python
args = {'Season': (2011, 2012),
       'PlayerName': ('BAILEY_JAMIL', 'BLAKE_CHRIS')}
args = [key+' IN ' +str(args[key]) for key in args]
where = ' AND '.join(args)

que = """
Select
    PlayerName 
from
    players
WHERE
    {where}



""".format(where=where)
print(que)

df = pd.read_sql(que, conn)
```
